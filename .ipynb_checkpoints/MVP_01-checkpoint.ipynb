```python
# Importando os módulos de manipulação e visualização de dados em python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# funções utilidades

# calcula percentual
def calc_percentual(lista_total, lista_secondaria):
    
    lista_resultado = []
    
    for i, k in zip(lista_total,lista_secondaria):
        lista_resultado.append(k / i * 100)
    return lista_resultado

# ordena valores
def ordena_lista(lista_labels, lista_valores):
    dicionario_dados = {}
    dicionario_ordenado = {}
    
    for chave, valor in zip(lista_labels, lista_valores):
        dicionario_dados[chave] = valor
     
    lista_chave = []
    lista_valores = []
    for i in sorted(dicionario_dados, key = dicionario_dados.get):  
        #dicionario_ordenado[i] = dicionario_dados[i]
        lista_chave.append(i)
        lista_valores.append(dicionario_dados[i])
    
    return lista_chave, lista_valores 
```

```python
# Criando uma nova coluna média dos dados, antes disso vamos criar uma copia do dataset original e trabalhar com a copia.

data_with_average = data
```

```python
# Vamos verificar agora quais os alunos obteveram as melhores média formada pelas três notas.
# Para isso, vou criar uma nova coluna com a média das três notas. 
# Vou utilizar a função 'round' para limitar a quanidade de casas decimais em 2.  

data_with_average['average score'] = round((data_with_average['writing score'] + 
                                      data_with_average['reading score'] + 
                                      data_with_average['math score']) / 3, 2)
```

```python
# Vamos verificar a propoção de alunos que tiraram média menor que 60 pontos. Vamos supor que 60 pontos é a nota mínima 
# para passar no teste. 

filtro_average_less = data_with_average['average score'] < 60  # filtro para selecionar alunos que tiraram menos que 60 pontos. 
temp_data = data_with_average.loc[filtro_average_less]

# Aqui estou extraíndo os totalizadores para apresentar em um gráfico de pizza.
total_geral_alunos = len(data_with_average)
total_alunos_reprovados = len(temp_data)
total_alunos_aprovados = total_geral_alunos - total_alunos_reprovados

# Configuração do gráfico
lista_propocoes = [total_alunos_aprovados,total_alunos_reprovados]
descricoes = ['ALUNOS APROVADOS','ALUNOS REPROVADOS']
destaque = [0, 0.1] # Parâmetro para destacar os alunos reprovados

plt.pie(lista_propocoes, labels=descricoes, explode=destaque, autopct='%.2f%%')
plt.show()
```

```python
# Verificando a propoção de alunos que tiraram média >= 60 pontos e que almoçaram normal x tiveram almoço reduzido. 

filtro_average_more = data_with_average['average score'] >= 60         # filtra alunos que tiraram média >= 60
descricoes = ['ALUNOS COM ALMOÇO NORMAL','ALUNOS COM ALMOÇO REDUZIDO']

df_alunos_aprovados = data_with_average.loc[filtro_average_more]

print('Quantidade de alunos por tipo de almoço.')
print(data_with_average['lunch'].value_counts())
print('------------------------------------------------------------------------------------')
print('Quantidade de alunos por tipo de almoço aprovado com média >= 60 pontos.')
print(df_alunos_aprovados['lunch'].value_counts())

#df_alunos_aprovados_lanche = df_alunos_aprovados.loc[filtro_lanche]
#df_alunos_aprovados_reduced_lanche = df_alunos_aprovados.loc[filtro_reduced_lanche]

#qtd_alunos_aprovados_reduced_lanche = len(df_alunos_aprovados_reduced_lanche)
#qtd_alunos_aprovados_lanche = len(df_alunos_aprovados_lanche)

#lista_propocoes = [qtd_alunos_aprovados_lanche, qtd_alunos_aprovados_reduced_lanche]

#destaque = [0, 0.1] # vai destacar os alunos reprovados
#plt.pie(lista_propocoes, labels=descricoes, explode=destaque, autopct='%.2f%%')
#plt.show()
```

```python
# Criando listas de valores e labels dos dados totais.
lista_labels_totais_lanch = data_with_average['lunch'].value_counts().keys()
lista_valores_totais_lanch = data_with_average['lunch'].value_counts().values

# Criando lista de valores e labels dos dados de alunos aprovados por média.
lista_labels_aprovados_lanch = df_alunos_aprovados['lunch'].value_counts().keys()
lista_valores_aprovados_lanch = df_alunos_aprovados['lunch'].value_counts().values

# listas após algumas operações
lista_valores_percentual_lanch = calc_percentual(lista_valores_totais_lanch, lista_valores_aprovados_lanch)
lista_chave_lanch, lista_valores_percentual_lanch = ordena_lista(lista_labels_aprovados_lanch, 
                                                                 lista_valores_percentual_lanch)
destaque = [0, 0.1] # vai destacar os alunos reprovados
plt.pie(lista_valores_percentual_lanch, labels=lista_chave_lanch, explode=destaque, autopct='%.2f%%')
plt.show()
```

<p> Levando em consideração o gráfico acima, almoçar normal contribuiu para passar no tete.</p>

### Vamos verificar a seguir quais grau de formação dos pais se destacou em relação aos alunos que passaram por média.

```python
# 1 Verificando a quantidade de ocorrência por grau de formação dos pais.
# 2 Verificando a quantidade de ocorrência por grau de formação dos pais dos alunos aprovados por média.

print('Quantidade de ocorrência por grau de formação dos pais.')
print(data_with_average['parental level of education'].value_counts())
print('------------------------------------------------------------------------------------')
print('Quantidade de ocorrência por grau de formação dos pais de alunos aprovados por média.')
print(df_alunos_aprovados['parental level of education'].value_counts())
```

```python
# Vamos agora verificar o percentual de aprovação dos alunos com base no grau de formação dos pais

# Criando listas de valores e labels dos dados totais.
lista_labels_total = data_with_average['parental level of education'].value_counts().keys()
lista_valures_total = data_with_average['parental level of education'].value_counts().values

# Criando lista de valores e labels dos dados de alunos aprovados por média.
lista_labels_aprovados = df_alunos_aprovados['parental level of education'].value_counts().keys()
lista_valures_aprovados = df_alunos_aprovados['parental level of education'].value_counts().values

# listas após algumas operações
lista_valures_percentual = calc_percentual(lista_valures_total, lista_valures_aprovados)
lista_chave, lista_valores_percentual = ordena_lista(lista_labels_aprovados, lista_valures_percentual)

lista_chave = lista_chave[::-1] # inverti a lista
lista_valores_percentual = lista_valores_percentual[::-1] # inverti a lista

# início da configuração do gráfico de barras.
fig = plt.figure(figsize = (15, 5))
plt.bar(lista_chave, lista_valores_percentual, width = 0.6)
plt.xlabel("Nível de escolaridade dos pais.")
plt.ylabel("% de alunos aprovados por média.")
plt.title("Percentual de alunos aprovados por média por escolaridade dos pais por ")
plt.show()
```

<p> Como podemos perceber a ordem dos melhores percentuais de aprovação dos alunos segue a ordem de maior escolaridade dos pais.</p>
